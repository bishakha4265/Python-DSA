HW1_1: From string s, create a list of strings L, each of which is of length 1

HW1_2: Create a function reverse_range for reversing a part of the string, represented by list L, defined by start and finish indices

```python
# HW1_2: write a function for reversing a part of the list
def reverse_range(L, start, finish):
  while(start<finish):
    L[start],L[finish]=L[finish],L[start]
    start=+1
    finish=-1
  return(L)
```

```python
# check for the expected output:
L_test=['t', 'h', 'e', ' ', 'b', 'o', 'o', 'k', ' ', 'i', 's', ' ', 'r', 'e', 'd']
reverse_range(L_test, 4, 7)
print(L_test)
# output:
# ['t', 'h', 'e', ' ', 'k', 'o', 'o', 'b', ' ', 'i', 's', ' ', 'r', 'e', 'd']
```

HW1_3: Create a function reverse_words for reversing words in a sentence that uses reverse_range function

```python
# HW1_3: reverse words in the sentence by first reversing the sentence as a whole and then reverse individual words using reverse_range function, use the above function within
def reverse_words(L):
  L_reverse=L[::-1]
  print(L_reverse)
  count=0
  start=0
  for c in L_reverse:
    if c != ' ':
      count=count+1
    if c == ' ' or count == len(L_reverse):
      reverse_range(L_reverse, start, count-1)
      count=count+1
      start=count
  return(L_reverse)



```

HW1_5: Not in-place operation (i.e. the function returns a value): Produce the same result and new string s1, applying string functions ( join, reversed, split) on s.

```python
def reverse_words_str(s):
  s_list=s.split()
  s_list.reverse()
  s1_sentence=' '.join(s_list)
  return(s1_sentence)
```
