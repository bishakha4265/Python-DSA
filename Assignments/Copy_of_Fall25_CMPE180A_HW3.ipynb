# HW3.1 Binary tree of random numbers

The following code illustrates the binary tree class **RBT**, its methods and use. Your task is to comment all code sections.

```python
# comment 1: Import random module to generate psuedo random numbers
# Tip: check official Python documentation
import random


# comment 2: Return a random floating-point number N such that a <= N <= b for a <= b
random.uniform(-1,1)
```

```python
# comment 3:
'''
Create RBTNode Class , initialize class obj with the left and right pointer pointing to None and value v. __repr__ function is used to define string
representation of the object.

'''
class RBTNode:
    def __init__(self,v):
        self.lpt=None
        self.rpt=None
        self.v=v
    def __repr__(self):
        return "{}".format(self.v)
```

```python
class RBT:
    # comment 4:
    ''' init function to intialize root with None when an object RBT Class is created '''
    def __init__(self):
        self.root=None

    # comment 5:
    '''
    rec_add funtion is used to add new_node to the tree. it is called from inside add_item function if root node is not None.
    if the value v of new_node is less than equal to tree_node(root/parent) then it goes into left pointer lpt of the root/parent,
      else the fucntion is recursively called till it finally finds correct position to add the new_node.
    similarly,
    if the value v of the new node is greater than equal to tree_node(root/parent) then it goes into right pointer rpt of the root/parent,
       else the fucntion is recursively called till it finally finds correct position to add the new_node.
    '''
    def rec_add(self,tree_node,new_node):
        if new_node.v <= tree_node.v:
            # less than, left
            if tree_node.lpt==None:
                tree_node.lpt=new_node
                return
            else:
                self.rec_add(tree_node.lpt,new_node)
        else:
            # greater than side
            if tree_node.rpt==None:
                tree_node.rpt=new_node
                return
            else:
                self.rec_add(tree_node.rpt,new_node)
    # comment 6:
    '''
    add_item function is used to add new_node to the tree , either as a root or as child using rec_add funtion.
    '''
    def add_item(self,rbt):
        rbt.rpt=None
        rbt.lpt=None
        if self.root==None:
            self.root=rbt
            return
        self.rec_add(self.root,rbt)
```

```python
# comment 7:
'''
Creates an object of RBT Class and intialize its root with None using init method.
'''
bt1=RBT()
```

```python
# comment 8:
'''
For loop runs from 50 times , creates object of class RBTNode with a random value between -1 and 1.
then the new node d is inserted into the tree bt1 using add_item function.

'''
for ix in range(50):
    d=RBTNode(random.uniform(-1,1))
    bt1.add_item(d)
```

```python
# comment 12:
'''
walktree function visits the left node , then root , then right node , which is inorder traversal of the tree.

'''
def walktree(nd):
    if nd==None:
        return
    walktree(nd.lpt)
    print(nd)
    walktree(nd.rpt)
```

3. Create one instance of WBT class and read in the file 'oceanDepths.words' and place the words in the instantiateed binary tree. While filling in the tree, keep track also of its **depth** value. Depth for the root node is 0, for the root's children nodes it is 1; in general: if the parent node has depth **i**, its children will have depth **i+1**. See the function signatures below.

4. Traverse the created tree, and print the words whose depth is the same as the maximum depth i.e. equal to the height of the tree.

The sceleton of the code and the signitures of the functions as well as the test code is given below. Comment and document all created functions and classes. Print out also the number of occurences of those words.



```python
# Binary Tree Node Class
class Node:
    def __init__(self, word):
      # add code #1:
      self.lpt=None
      self.rpt=None
      self.word=word
      self.depth=0
      self.count=0

    def __str__(self):
        return f"{self.word} (count: {self.count}, depth: {self.depth})"

class WBT:

  def __init__(self):
    # add code #2:
    self.root=None
    self.max_depth = 0

  # Function to insert a word into the binary tree
  def insert(self, tree_node, new_node, depth):
   # add code #3: (tip: look at  rec_add and add_item functions from M3_1)
        if new_node.word == tree_node.word:
            tree_node.count += 1
            return

        # Decide to go left or right
        if new_node.word <= tree_node.word:
            if tree_node.lpt is None:
                new_node.depth = depth + 1
                tree_node.lpt = new_node
                new_node.count += 1
                self.max_depth = max(self.max_depth, new_node.depth)
            else:
                self.insert(tree_node.lpt, new_node, depth + 1)
        else:
            if tree_node.rpt is None:
                new_node.depth = depth + 1
                tree_node.rpt = new_node
                new_node.count += 1
                self.max_depth = max(self.max_depth, new_node.depth)
            else:
                self.insert(tree_node.rpt, new_node, depth + 1)

  def add_item(self, word):
        if len(word) <= 3:
            return  # ignore short words

        new_node = Node(word)
        if self.root is None:
            self.root = new_node
            self.max_depth = 0
        else:
            self.insert(self.root, new_node, 0)


  # Traverse the binary tree and print words at the maximum depth
  def print_max_depth_words(self, root, max_depth):
      # add code #4:
      # tip: check if root is None; print if depth is max; recursevely
      # call function for both child nodes if they are different from None
      if root is None:
        return
      self.print_max_depth_words(root.lpt, max_depth)
      if root.depth == max_depth:
         print(root)
      self.print_max_depth_words(root.rpt, max_depth)



  # Create a binary tree from the file with the name filename
  def create_binary_tree(self, filename):
    # add code #5
    root = None
    max_depth = 0

    # mount the google drive (see tutorial 2) and set the filename path
    filename = "/content/sample_data/oceanDepths.words" # fill in
    # open the file and start reading word by word from it
    with open(filename, "r") as file:
        words = file.read().splitlines()

    # create and populate the tree
    for word in words:
      # Start insertion from the root
      # Update maximum depth : max_depth
      word = word.strip()
      self.add_item(word)
      max_depth = self.max_depth

    return root, max_depth
```
