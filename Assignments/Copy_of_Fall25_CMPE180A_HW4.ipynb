# HW4.1 Histogram

Create histograms for the two data sets using a suitable binning interval and plot them on the same figure.

Questions:

*   how many heights are provided in each file and in how many rows?
*   Calculate and print out the mean and standard deviation for each sex.



```python
import numpy as np
import matplotlib.pyplot as plt

# Load the heights
male_heights = np.loadtxt('/content/drive/MyDrive/Assign-4/ex6-3-f-male-heights.txt')

female_heights = np.loadtxt('/content/drive/MyDrive/Assign-4/ex6-3-f-female-heights.txt')

print("Male data shape:", male_heights.shape)
print("Female data shape:", female_heights.shape)
male_heights = male_heights.flatten()
female_heights = female_heights.flatten()

num_male_values = male_heights.size
num_male_rows = np.loadtxt('/content/drive/MyDrive/Assign-4/ex6-3-f-male-heights.txt').shape[0]
num_female_values = female_heights.size
num_female_rows = np.loadtxt('/content/drive/MyDrive/Assign-4/ex6-3-f-female-heights.txt').shape[0]
print(f"Male: {num_male_values} heights in {num_male_rows} rows")
print(f"Female: {num_female_values} heights in {num_female_rows} rows")


male_mean = np.mean(male_heights)
male_std = np.std(male_heights)
female_mean = np.mean(female_heights)
female_std = np.std(female_heights)

print(f"Male mean height: {male_mean:.2f} cm, std dev: {male_std:.2f} cm")
print(f"Female mean height: {female_mean:.2f} cm, std dev: {female_std:.2f} cm")

plt.figure(figsize=(10, 6))

min_height = min(male_heights.min(), female_heights.min())
max_height = max(male_heights.max(), female_heights.max())
bins = np.arange(min_height, max_height + 1, 1)  # 1 cm intervals

plt.hist(male_heights, bins=bins, alpha=0.6, label='Male', color='blue', edgecolor='black')
plt.hist(female_heights, bins=bins, alpha=0.6, label='Female', color='pink', edgecolor='black')

plt.title('Height Distribution by Sex')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

# HW4.2 Histogram in Imperial Units

Repeat the exercise in imperial units (feet and inches).

```python

male_heights_in = male_heights / 2.54
female_heights_in = female_heights / 2.54

num_male_values = male_heights_in.size
num_male_rows = np.loadtxt('/content/drive/MyDrive/Assign-4/ex6-3-f-male-heights.txt').shape[0]

num_female_values = female_heights_in.size
num_female_rows = np.loadtxt('/content/drive/MyDrive/Assign-4/ex6-3-f-female-heights.txt').shape[0]

print(f"Male: {num_male_values} heights in {num_male_rows} rows")
print(f"Female: {num_female_values} heights in {num_female_rows} rows")
male_mean_in = np.mean(male_heights_in)
male_std_in = np.std(male_heights_in)

female_mean_in = np.mean(female_heights_in)
female_std_in = np.std(female_heights_in)

def inches_to_feet_and_inches(inches):
    feet = int(inches // 12)
    rem_inches = inches % 12
    return feet, rem_inches

# Convert means to feet + inches
male_feet, male_rem_inches = inches_to_feet_and_inches(male_mean_in)
female_feet, female_rem_inches = inches_to_feet_and_inches(female_mean_in)

print(f"Male mean height: {male_mean_in:.2f}\" ({male_feet} ft {male_rem_inches:.1f} in), std dev: {male_std_in:.2f}\"")
print(f"Female mean height: {female_mean_in:.2f}\" ({female_feet} ft {female_rem_inches:.1f} in), std dev: {female_std_in:.2f}\"")

# Create histogram in inches
plt.figure(figsize=(10, 6))

# Use bin size of 1 inch
min_height = min(male_heights_in.min(), female_heights_in.min())
max_height = max(male_heights_in.max(), female_heights_in.max())
bins = np.arange(np.floor(min_height), np.ceil(max_height) + 1, 1)  # 1 inch bins

plt.hist(male_heights_in, bins=bins, alpha=0.6, label='Male', color='blue', edgecolor='black')
plt.hist(female_heights_in, bins=bins, alpha=0.6, label='Female', color='pink', edgecolor='black')

plt.title('Height Distribution by Sex (inches)')
plt.xlabel('Height (inches)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
